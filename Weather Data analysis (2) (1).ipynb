```python
# Renaming columns for consistency
wd.rename(columns={'Rel Hum_%': 'Humidity', 'Weather': 'Weather_condition'}, inplace=True)
```

```python
# Data Exploration
print("\nNumber of unique values in each column:")
print(wd.nunique())
print("\nUnique weather conditions:")
print(wd['Weather_condition'].unique())
```

```python
# Check for missing values in each column
missing_values_per_column = wd.isnull().sum()

# Check if there are any missing values in the entire DataFrame
any_missing_values = wd.isnull().any().any()

# Print the results
print("Missing values per column:")
print(missing_values_per_column)

print("\nAre there any missing values in the DataFrame? ", any_missing_values)
```

```python
# Data Exploration
print("\nNumber of unique values in each column:")
print(wd.nunique())
print("\nUnique weather conditions:")
print(wd['Weather_condition'].unique())
```

```python
# Show all records where weather condition is 'Clear'
print("\nRecords where Weather Condition is Clear:")
print(wd[wd['Weather_condition'] == 'Clear'])
```

```python
plt.figure(figsize=(8, 6))
plt.hist(wd['Humidity'], bins=20, color='green', alpha=0.7)
plt.xlabel('Humidity')
plt.ylabel('Frequency')
plt.title('Humidity Distribution')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Temp_C', y='Dew Point Temp_C', data=wd, color='red')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Dew Point Temperature (Â°C)')
plt.title('Temperature vs. Dew Point Temperature')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=wd['Wind Speed_km/h'], color='orange')
plt.xlabel('Wind Speed (km/h)')
plt.title('Wind Speed Distribution')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Weather_condition', data=wd, palette='viridis')
plt.xlabel('Weather Condition')
plt.ylabel('Count')
plt.title('Weather Condition Counts')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Plot a histogram of wind speeds
plt.figure(figsize=(10, 6))
sns.histplot(wd['Wind Speed_km/h'], bins=20, color='skyblue', kde=True)
plt.title('Distribution of Wind Speeds')
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.lineplot(data=wd, x='Date/Time', y='Temp_C')
plt.title('Temperature Variation Over Time')
plt.xlabel('Date/Time')
plt.ylabel('Temperature (Celsius)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
